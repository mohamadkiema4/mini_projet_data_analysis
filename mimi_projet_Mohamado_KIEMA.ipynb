## <span style='color:blue'> Importation des données </span> 

```python
import pandas as pd

# Ouverture du fichier texte et lecture de son contenu
with open('BF.txt', 'r',encoding='utf-8') as fichier:
    lignes = fichier.readlines()

# Suppressions des sauts de lignes 
donnees = [ligne.strip().split('\t') for ligne in lignes]  

# Créer un DataFrame à partir des données
df = pd.DataFrame(donnees)

# Afficher le DataFrame
df
```

## <span style='color:blue'> 3-Opérations de prétraitement</span>

## <span style='color:blue'> 3-1) Identification de colonnes</span>

```python
# Conservation des colonnes:  Identifiants,Noms de lieux, latitudes et longitudes
# Dans le read me ces colonnes correspond aux valeurs suivantes: 0, 1,4 et 5
# la colonne 0 correspond à l'identifiant entier de l'enregistrement dans la base de données de noms géographiques
# la colonne 1 est le nom du point géographique
# la colonne 4 correspond à la latitude
# la colonne 5 correspond à la longitude
df_filtr=df[[0,1,4,5]]
df_filtr
```

### <span style='color:blue'> 3-2) Renommer des colonnes</span>

```python
# Renommons Ces colonnes
df_filtr= df_filtr.rename(columns={0: 'Id', 1: 'location_name', 4: 'lat', 5: 'long'})
df_filtr
```

### <span style='color:blue'> 3-3) Sauvegarde de fichier sous format csv</span>

```python
# Sauvegarde de ces données dans un fichier CSV
df_filtr.to_csv('burkina_location.csv', index=False)
```

## <span style='color:blue'> 4) Opérations sur le fichier CSV burkina_location.csv.</span>

```python
# importation des données
df_bf=pd.read_csv('burkina_location.csv')
df_bf
```

### <span style='color:blue'> 4-1)-Extraction des données contenant le nom 'gounghin', enregistrez-le sous le fichier gounghin.csv </span>

```python
# extraction des données contenant le nom 'gounghin'
df_goung=df_bf[df_bf['location_name']=='Gounghin']
df_goung
```

```python
# Sauvegarde du data frame sous format csv
df_goung.to_csv('gounghin.csv', index=False)
```

### <span style='color:blue'> 4-2) Extraction des noms de lieux dont les premières lettres  sont compris entre 'A' et 'P' (ordre alphabétique)</span>

```python
# sélection des noms de lieux compris entre 'A' et 'P'
lieux_AP = df_bf[df_bf['location_name'].str[0].between('A', 'P')]
# Trier le DataFrame résultant par ordre alphabétique des noms de lieux
lieux_AP=lieux_AP.sort_values(by='location_name')

# Réinitialiser les index après le tri
lieux_AP = lieux_AP.reset_index(drop=True)
lieux_AP
```

```python
lieux_AP.to_csv('lieux.csv',index=False)
```

### <span style='color:blue'> 4-3) Identification respective, de la latitude, de la longitude minimale et les noms de lieux correspondants </span>

#### <span style='color:blue'> Latitude minimale</span>

```python
# Reccherche de la latitude minimale et le nom du lieu correspondant
# trions le dataframe par ordre croissant d'abord
df_bf.sort_values(by = 'lat')
```

```python
# La latitude minimale correspond à la première ligne du dataframe ci-dessus et son index est 6018
# Affichons le nom du lieu, ID et la latitude minimale
lat_min=df_bf.iloc[6018, :]
lat_min
```

#### <span style='color:blue'> Longitude minimale </span>

```python
# Reccherche de la longitude minimale et le nom du lieu correspondant
# trions le dataframe par ordre croissant d'abord
df_bf.sort_values(by = 'long')
```

```python
# La longitude minimale correspond à la première ligne du dataframe ci-dessus et son index est 4221
# Affichons le nom du lieu, ID et la longitude minimale
long_min=df_bf.iloc[4221, :]
long_min
```

### <span style='color:blue'> 4-4) Les lieux dont les coordonnées sont comprises entre (lat >= 11 et lon <= 0.5).</span>

```python
# Sélectionner les lieux selon les critères spécifiés
lieux_selectionnes = df_bf.loc[(df_bf['lat'] >= 11) & (df_bf['long'] <= 0.5)]
lieux_selectionnes
```

```python
# les noms de ces lieux
lieux_selectionnes['location_name']
```

### <span style='color:blue'> 5-2) Créer une feuille dans ce fichier, du nom gounghin et enregistrer les données contenant le nom 'gounghin' obtenues dans 4.1 </span>

```python
# Importation d'une méthode nécéssaire
from openpyxl import load_workbook

# Écrire les données "gounghin" dans la nouvelle feuille Excel
with pd.ExcelWriter('mini_projet.xlsx', engine='openpyxl', mode='a') as writer:
    writer.book = load_workbook('mini_projet.xlsx')
    df_goung.to_excel(writer, sheet_name='gounghin', index=False)
    writer.save()
```

### <span style='color:blue'> 5-3)Créer une second feuille dans ce même fichier, du nom A_to_P et enregistrer les données de 4.2</span>

```python
# Écrire les données "gounghin" dans la nouvelle feuille Excel
with pd.ExcelWriter('mini_projet.xlsx', engine='openpyxl', mode='a') as writer:
    writer.book = load_workbook('mini_projet.xlsx')
    lieux_AP.to_excel(writer, sheet_name='A_to_P', index=False)
    writer.save()
```
